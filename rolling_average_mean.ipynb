```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('walmart_stock.csv', index_col = 'Date', parse_dates = True)
```

```python
df.head()
```

```python
df['Open'].plot(figsize=(16,6))
```

```python
#Bollengers Band

#Close 20 MA 
df['Close: 20 Day Mean'] = df['Close'].rolling(20).mean()
# Upper = 20MA + 2*std(20)
df['Upper'] = df['Close: 20 Day Mean'] + 2*(df['Close'].rolling(20).std())
# lower = 20MA - 2*std(20)
df['Lower'] = df['Close: 20 Day Mean'] - 2*(df['Close'].rolling(20).std())
#Close
df[['Close', 'Close: 20 Day Mean', 'Upper', 'Lower']].plot(figsize=(16,6))
```

```python
#to see amount of days you can use soemthing like this to see last 100 days
#Close 20 MA 
df['Close: 20 Day Mean'] = df['Close'].rolling(20).mean()
# Upper = 20MA + 2*std(20)
df['Upper'] = df['Close: 20 Day Mean'] + 2*(df['Close'].rolling(20).std())
# lower = 20MA - 2*std(20)
df['Lower'] = df['Close: 20 Day Mean'] - 2*(df['Close'].rolling(20).std())
#Close
df[['Close', 'Close: 20 Day Mean', 'Upper', 'Lower']].tail(100).plot(figsize=(16,6))
```
